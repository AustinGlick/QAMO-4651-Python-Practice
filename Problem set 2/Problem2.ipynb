```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df = pd.read_stata('CEOSAL2.DTA')
df.describe()
```

```python
df[df.ceoten == 0].shape
#Total of 5 ceos at 0 years 
```

```python
model = smf.ols(formula='lsalary ~ ceoten', data=df)
results = model.fit()
print(results.summary())
```
